```python
age_range = [17,22,27,35,200]
age_labels = ['Young Adults', 'Early Adults', 'Adults', 'Middle-aged and Above']

education_range = [-1,12,15,200]
education_labels = ['Primary Education', 'Secondary Education', 'Higher Education']

income_range = [-1,40000,60000,80000,10000000000]
income_labels = ['Low Income', 'Moderate Income', 'High Income', 'Very High Income']

miles_range = [-1,50,100,200,1000]
miles_labels = ['Light', 'Moderate', 'Active', 'Extreme']
```

```python
df['age_group'] = pd.cut(df['Age'], bins = age_range, labels= age_labels)

df['education_group'] = pd.cut(df['Education'], bins = education_range, labels= education_labels)

df['income_group'] = pd.cut(df['Income'], bins = income_range, labels= income_labels)

df['miles_group'] = pd.cut(df['Miles'], bins = miles_range, labels= miles_labels)
```

```python
ax = sns.countplot(x=df['Product'], order=df['Product'].value_counts(ascending=False).index);

abs_values = df['Product'].value_counts(ascending=False).values

ax.bar_label(container=ax.containers[0], labels=abs_values)
plt.show()
```

```python
## adding price column.

price_map = {
    'KP281': 1500,
    'KP481': 1750,
    'KP781': 2500
}

df['price'] = df['Product'].map(price_map)
```

```python
sales = df.groupby('Product')['price'].agg(np.sum)
sales
```

```python
fig = plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
bars = plt.bar(sales.index, sales.values, color='forestgreen')
for bar, value in zip(bars, sales.values):
  plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height(),
             value,
             ha='center',
             va='bottom')

plt.xlabel('Product', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.title('Product wise sales', fontsize=18)

plt.subplot(1,2,2)
ax = sns.countplot(x=df['Product'], order=df['Product'].value_counts(ascending=False).index);
abs_values = df['Product'].value_counts(ascending=False).values
ax.bar_label(container=ax.containers[0], labels=abs_values)

plt.show()
```

Above graphs show that Product KP281 have highest sales with respect to price as well as quantity, KP481 being on second and KP781 on third.
